#### __Importing libraries to analyze data__
  ---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class modified_redwine:
  
    def __init__(self,data):
        self.data=pd.read_csv(data)
        
    def fill_null(self):
        return self.data.fillna(self.data.mean(), inplace = True, )
       
    def remove_duplicates(self):
        
        return self.data.drop_duplicates(inplace=True)
    
    def preview_head(self):
        return self.data.head(10)

    def preview_tail(self):
        return self.data.tail(20)

    def check_info(self):
        return self.data.info()

    def data_description(self):
        return self.data.describe()

    def barplot(self):
        
        plt.figure(figsize=(10,8))
        column1= input ('Enter column name of x-asis: ')
        column2= input ('Enter column name of y-axis:  ')
        colour= input('Enter  colour of choice:   ')
        plt.bar(self.data[column1],self.data[column2], width=0.18, color=(colour), edgecolor= 'black', linewidth=1)
        plt.xlabel(column1)
        plt.ylabel(column2)
        plt.title(f' Bar Graph of {column2} Against {column1}') 
        plt.grid(True)
        plt.show()

    def hist(self):
        plt.figure(figsize=(10, 8))
        column= input('Enter column name:  ' )
        colour= input('Enter  colour of choice:   ')
        plt.hist(self.data[column], bins=20, color=(colour), edgecolor='black')
        plt.xlabel(column)
        plt.title(f'Histogram of {column}')
        plt.grid(True)
        plt.show()
        
    
    def scatterplot(self):
        plt.figure(figsize=(10,8))
        column1= input ('Enter column name of x-asis: ')
        column2= input ('Enter column name of y-axis:  ')
        colour= input('Enter  colour of choice:   ')
        plt.scatter(self.data[column1],self.data[column2], color=(colour), alpha=0.5)
        plt.title(f'Scatter Diagram of {column2} Against {column1}')
        plt.xlabel(column1)
        plt.ylabel(column2)
        plt.grid(True)
        plt.show()

    def pie_chart(self):
        quality_counts= self.data['quality'].value_counts()
        plt.figure(figsize=(8,8))
        plt.pie(quality_counts, autopct='%1.1f%%', startangle=90)
        plt.legend(quality_counts.index, title ='QUALITY', loc='upper right')
        plt.title('Distribution of Wine Quality')
        plt.show()

    def correlation_plot(self):
        correlation = self.data.corr()

        plt.figure(figsize=(10, 8))
        plt.imshow(correlation, cmap='coolwarm', interpolation='nearest')
        plt.colorbar()
        plt.title('Correlation Plot')
        plt.xticks(range(len(correlation)),correlation.columns, rotation=90)
        plt.yticks(range(len(correlation)),correlation.columns)
        plt.tight_layout()
        plt.show()

    
dataset = modified_redwine('red_wine.csv')   
```
